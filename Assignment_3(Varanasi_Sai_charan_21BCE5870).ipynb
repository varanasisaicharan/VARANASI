6. Find the outliers and replace them outliers

```python
for j in x.columns[2:6]:
  q1 = x[j].quantile(0.25) #Q1
  q3 = x[j].quantile(0.75) #Q3
  print("\nQ1",j,":",q1)
  print("Q3",j,":",q3)
  IQR=q3-q1
  print("IQR",j,":",IQR)
  upper_limit = q3+1.5*IQR
  print("Upper limit",j,":",upper_limit)
  lower_limit =q1-1.5*IQR
  print("Lower limit",j,":",lower_limit)

```

```python
x.median()
```

```python
for j in x.columns[2:5]:
  x[j] = np.where(x[j]>upper_limit,30,x[j])
```
